```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Specify the path to your data file
data_file_path = '/Users/bhanuprasadthota/trapezoidal_rule_errors.txt'

# Read data from the specified file path
data = np.genfromtxt(data_file_path, names=['N', 'error'], delimiter=' ')
```

```python
# Calculate log2(N) and log2(error)
log2_N = np.log(data['N'])
log2_error = np.log(data['error'])
```

```python
# Plot the log-log graph
plt.figure(figsize=(10, 6))
plt.plot(log2_N, log2_error, marker='o', linestyle='-', color='b')
plt.title('Log-Log Plot of log2(N) vs. log2(Error)')
plt.xlabel('log2(N)')
plt.ylabel('log2(Error)')
plt.grid(True)

plt.show()
```

```python
from scipy.stats import linregress

# Fit a linear regression line
slope, intercept, r_value, p_value, std_err = linregress(log2_N, log2_error)

# Print the calculated slope
print(f"Slope (C) of the log-log line (with log2): {slope:.4f}")
```
